```python
import pandas as pd
import warnings
```

```python
warnings.filterwarnings("ignore")
```

```python
df = pd.read_excel("Dataset.xlsx")
```

```python
dataset = pd.DataFrame(df)
```

```python
dataset.head(2)
```

```python
dataset.info()
```

```python
missing_values = dataset.isnull().sum()
missing_values = missing_values[missing_values > 0]
```

```python
missing_values
```

```python
dataset["amount"].fillna("0", inplace=True)
```

```python
dataset["mode_collection_cash_upi_banktransfer"].fillna("NA", inplace=True)
```

```python
dataset["farmer_payment_date"] = pd.to_datetime(dataset["farmer_payment_date"])
```

```python
dataset["farmer_payment_date"].fillna(method="ffill", inplace=True)
```

```python
dataset["plantation_date"] = pd.to_datetime(dataset["plantation_date"])
```

```python
dataset["plantation_date"].fillna(method="bfill", inplace=True)
```

```python
dataset.columns = dataset.columns.str.lower().str.replace(" ", "_")
```

```python
categorical_cols = ["farmer_name", "district", "block", "water_available", "electricity_available"]
for col in categorical_cols:
    dataset[col] = dataset[col].astype(str).str.title()
```

```python
numeric_cols = ["total_land_area_acre", "area_f4f_acre"]
dataset[numeric_cols] = dataset[numeric_cols].apply(pd.to_numeric, errors="coerce")
```

```python
dataset["area_f4f_acre"] = df.apply(
    lambda row: row["total_land_area_acre"] if row["area_f4f_acre"] > row["total_land_area_acre"] else row["area_f4f_acre"],
    axis=1
)
```

```python
dataset = dataset.drop_duplicates(subset=["uid"], keep="first")
```

```python
invalid_water_values = dataset[~dataset["water_available"].isin(["Yes", "No"])]
invalid_electricity_values = dataset[~dataset["electricity_available"].isin(["Yes", "No"])]
```

```python
cleaning_summary = {
    "Missing Values Fixed": missing_values.to_dict(),
    "Duplicate Entries Removed": dataset.shape[0] - dataset.shape[0],
    "Invalid Water Availability Values": invalid_water_values.shape[0],
    "Invalid Electricity Availability Values": invalid_electricity_values.shape[0],
    "Corrected Area Inconsistencies": (dataset["area_f4f_acre"] > dataset["total_land_area_acre"]).sum(),
}
```

```python
print(cleaning_summary)
```

```python
dataset.to_excel("data_cleaned.xlsx", index=False)
```
